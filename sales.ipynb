```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

plt.style.use('ggplot')


# reading data
df = pd.read_csv(r'csv_files/Sales.csv')
```

```python
# getting first 6 rows
df.head(n=6)
```

```python
# looking for missing data
for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print(pct_missing)
```

```python
df.dtypes
```

```python
df.sort_values(by=['Unit_Price'], inplace=False, ascending=False)
```

```python
# showing duplicates
df['Product'].drop_duplicates()
```

```python
# scutter plot

plt.scatter(x=df['Cost'], y=df['Profit'])
plt.xlabel("Cost")
plt.ylabel("Price")
```

```python
sns.regplot(x=df['Cost'], y=df['Profit'], data=df, scatter_kws={'alpha': 0.5, 'color': 'red'}, line_kws={'color': 'blue'})
```

```python
correlation_matrix = df.corr(numeric_only=True)
sns.heatmap(correlation_matrix, annot = True)

plt.show()
```
